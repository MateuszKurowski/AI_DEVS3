```python
from dotenv import load_dotenv

load_dotenv()
```

```python
import os
import json
from utilities.GetTaks import get_task

ai_devs_key = os.getenv('ai_devs_key')
response = get_task(f'data/{ai_devs_key}/json.txt')

data = response.json()
print(json.dumps(data, indent=4))
```

```python
data["apikey"] = ai_devs_key
```

```python
from utilities.OpenAiService import OpenAiService

ai_service = OpenAiService()
```

```python
for item in data["test-data"]:
    question = item["question"]
    if "+" in question:
        num1, num2 = map(int, question.split(" + "))
        correct_answer = num1 + num2
        
        if item["answer"] != correct_answer:
            item["answer"] = correct_answer
    
    if "test" in item:
        test_question = item["test"]["q"]
        answer = ai_service.ask(test_question)

        item["test"]["a"] = answer
```
